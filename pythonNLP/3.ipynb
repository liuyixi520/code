## 使用Unicode进行文字处理
- unicode支持一百万种以上的字符表达
- 编码点：用四位十六进制数表示一个字符
- 编码：将文本中的固有编码格式翻译成unicode编码，放入到内存中
- 解码：将unicode编码从内存中读出来，翻译成固有格式，放入到文本或者终端中

## 使用正则表达式检测词组搭配

```python
import re
import nltk
wordlist = [w for w in nltk.corpus.words.words('en') if w.islower()]
wordlist
```
