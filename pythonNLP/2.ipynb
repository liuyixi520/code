```python
from nltk.corpus import gutenberg
for fieid in gutenberg.fileids():
    num_chars = len(gutenberg.raw(fieid))
    num_words = len(gutenberg.words(fieid))
    num_sents = len(gutenberg.sents(fieid))
    num_vocab = len(set([w.lower() for w in gutenberg.words(fieid)]))
    print(fieid, num_chars, num_words, num_sents, num_vocab)
```

```python
# 布朗语料库
from nltk.corpus import brown
for fieid in brown.fileids():
    num_chars = len(brown.raw(fieid))
    num_words = len(brown.words(fieid))
    num_sents = len(brown.sents(fieid))
    num_vocab = len(set([w.lower() for w in brown.words(fieid)]))
    print(fieid, num_chars, num_words, num_sents, num_vocab)
```

```python
# 查看布朗语料库中的类别信息
brown.categories()
```

```python
brown_news = brown.words(categories='news')
fdist = nltk.FreqDist([w.lower() for w in brown_news])
fdist.most_common(20)
```

## 更多关于Python: 代码重用

- 函数
- 使用上面的两个技巧可以使得代码重用更加容易

```python
# 定义一个把单词变成复数的函数
def plural(word):
    if word.endswith('y'):
        return word[:-1] + 'ies'
    elif word[-1] in 'sx' or word[-2:] in ['sh', 'ch']:
        return word + 'es'
    elif word.endswith('an'):
        return word[:-2] + 'en'
    else:
        return word + 's'
# 测试上面的函数
print(f'plural("tree"): {plural("tree")}')
print(f'plural("fairy"): {plural("fairy")}')
print(f'plural("woman"):{plural("woman")}')
```
