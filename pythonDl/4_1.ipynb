# 机器学习基础
机器学习大致可以分为以下四类：
- 监督学习：常见的分类和回归问题
- 无监督学习：常见的聚类问题
- 自监督学习：常见的推理问题
- 强化学习：常见的搜索问题
## 评估机器学习模型
- 机器学习的目的是为了提升泛化能力，所以为了找到一个好的机器学习模型，我们常常需要在优化和泛化之间找到一个比较好的平衡点。
- 为了比较好的评估模型，我们常常将数据集划分成三个部分：训练集、验证集和测试集。
- 评估模型的时候要注意以下几个方面：
  - 数据要有代表性
  - 时间箭头
  - 数据冗余
## 数据预处理
数据预处理一般要进行以下几步：
- 数据集向量化：张量是机器学习和深度学习的数据载体，数据集向量化是将数据集转换成张量的过程。任何形式的数据：结构化数据、文本、图像、音频等都可以转换成张量。
- 数据标准化：将数据集转换成均值为0，方差为1的数据集。
- 缺失值处理：因为机器学习在学习参数的时候是对缺失值敏感的，所以在训练之前我们常常把缺失值通过均值填充或者删除。
## 特征工程
定义：通过更加简单的方式来描述问题，从而使得问题变得更加简单。

在深度学习之前，SVM和MLP等一些经典的机器学习模型都需要高度依赖特征工程，才可以更好的解决问题。但是深度学习中提出了卷积和池化的方法，就可以让它们去更好的做以前特征工程中的事情。让问题的解决变得更加简单。
## 欠拟合&过拟合
- 欠拟合和过拟合是机器学习中的两个常见问题：造成这两个的问题原因主要是因为模型的复杂度和数据的复杂度不匹配造成的。当数据的复杂程度较高时，就会发生欠拟合，当数据的复杂程度较低时，就会发生过拟合。
- 一般来说当模型的参数越多时(更深更广)，模型就越复杂。由于目前算力越来越强，所以发生过拟合的概率越来越高。抑制过拟合的手段通常有以下几种方式：
  - 降低模型的复杂度，用更浅的层，更窄的网络
  - 添加权重正则化
  - 添加Dropout正则化
## 机器学习的一般流程
- 定义问题、定义数据集
- 选择衡量模型的指标->损失函数
- 确定评估方法
- 准备数据
- 开发比基准模型更好的模型
- 扩大模型规模：开发过拟合的模型
- 模型正则化与超参数调节
